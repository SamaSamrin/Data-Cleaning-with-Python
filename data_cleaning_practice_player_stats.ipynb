```python
import pandas as pd
```

```python
df = pd.read_csv("player_stats.csv")
```

```python
df.head()
```

```python
df.info()
```

# Renaming columns for better understanding

```python
df = df.rename(columns={ "Mat" : "Matches", "Inns" : "Innings", "NO" : "Not Out", "HS" : "Highest Score", "Ave" : "Average", "BF" : "Balls Faced", "SR" : "Strike Rate" })
```

```python
df.head()
```

# Splitting Player column into 2 : Player_Name & Country

```python
df[['Player_Name', 'Country']] = df['Player'].str.split("(", n=1, expand=True)
```

```python
df.head()
```

```python
df = df.drop(columns = ["Player"])

df.head()
```

```python
df['Country']  = df['Country'].str.rstrip(")")

df.head()
```

# Reordering columns to bring the Player_Name column to the front

```python
temp_column_holder = df.pop('Player_Name')

df.insert(0, 'Player', temp_column_holder)

df.head()
```

# Splitting Span into 2 columns (Debut_Year & Final_Year)

```python
df[['Debut_Year', 'Final_Year']] = df['Span'].str.split('-', n=1, expand=True)

df.head()
```

```python
df = df.drop(columns= ['Span'])

df.head()
```

# Removing undesired or unnecessary characters

```python
#checking how the dataframe looks right now first

df.head(15)
```

```python
df["Highest Score"] = df["Highest Score"].str.rstrip('*')

df.head(10)
```

```python
df['4s'] = df['4s'].str.rstrip('+')

df.head(10)
```

```python
df['6s'] = df['6s'].str.rstrip("+")

df.head(10)
```

# Removing an invalid character present throughout the dataset

```python
df = df.replace("", pd.NA)

df.head(10)
```

# Checking the cleaned dataset so far

```python
df.info()
```

# Fill all the null values with 0 (for now)

```python
df = df.fillna(0)

df.head(10)
```

```python
df.info()
```

# Changing column data types into appropriate ones

```python
df = df.astype( {'Highest Score' : int,
                'Balls Faced' : int,
                '0' : int,
                '4s' : int,
                '6s' : int,
                'Debut_Year' : int,
                'Final_Year' : int})
```
