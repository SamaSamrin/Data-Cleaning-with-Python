# Renaming columns

```python
df.columns
```

```python
df = df.rename(columns= {'Mat' : 'Matches',
                    'NO': ' Not_Out',
                    'HS' : 'Highest_Score',
                    'Ave' : 'Average',
                    'BF' : 'Balls_Faced',
                    'SR' : 'Strike_Rate'})

df
```

# Separate Player column into Player_Name and Country

```python
df[['Player_Name', 'Country']] = df['Player'].str.split('(', n=1, expand=True)

df
```

```python
df['Country'] = df['Country'].str.replace(")", "")

df
```

## Dropping the Player column since we don't need it anymore

```python
df = df.drop(columns='Player')
```

```python
df
```

## Reordering the Player_Name column so that it comes first in the table

```python
temp_column_header = df.pop('Player_Name')
```

```python
df.insert(0, 'Player_Name', temp_column_header)
```

```python
df
```

# Separating the Span column into Rookie_Year and Final_Year

```python
df[['Rookie_Year', 'Final_Year']] = df['Span'].str.split('-', n=1, expand=True)

df = df.drop(columns='Span')

df
```

# Removing unwanted characters

```python
df['Highest_Score'] = df['Highest_Score'].str.strip('*')

df
```

```python
df = df.replace(r'[+*-/]', '', regex=True)

df
```

```python
df['Country'] = df['Country'].str.replace("ICC", "")

df
```

```python
df[df['Country']=='SA']
```

```python
df.sample(10)
```

# Checking for empty or null values and handling them

```python
df.info()
```

```python
df = df.replace("", pd.NA)
```

```python
df.info()
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df = df.fillna(0)
```

```python
df.isna().sum().sort_values(ascending=False)
```
