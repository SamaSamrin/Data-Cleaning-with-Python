```python
import pandas as pd
```

```python
df = pd.read_csv("player_stats.csv")

df
```

```python
df.describe()
```

```python
df.info()
```

```python
df.columns
```

```python
df = df.rename(columns={'Mat' : 'Matches',
                       'Inns' : 'Innings',
                        'NO' : 'Not_Out',
                        'HS' : 'Highest_Score',
                        'BF' : 'Balls_Faced',
                        'Ave' : 'Average',
                        'SR' : 'Strike_Rate'})

df
```

```python
df.columns
```

# Divide Span into Debut_Year and Final_Year

```python
df[['Debut_Year', 'Final_Year']] = df['Span'].str.split('-', n=1, expand=True)

df
```

## Deleting the redundant 'Span' column

```python
df = df.drop(columns='Span')

df
```

# Split Player into Player_Name and Country

```python
df[['Player_Name', 'Country']] = df['Player'].str.split('(', n=1, expand=True)

df
```

```python
df[['Player', 'Player_Name', 'Country']]
```

```python
df = df.drop(columns='Player')

df.columns
```

```python
df['Country'] = df['Country'].str.rstrip(')')

df['Country']
```

```python
df['Country'] = df['Country'].str.strip('ICC/')

df['Country']
```

```python
df
```

```python
temp_col_holder = df.pop('Player_Name')

df.insert(0, 'Player_Name', temp_col_holder)

df
```

```python
df.columns
```

# Deleting unwanted or invalid characters

```python
df
```

```python
df = df.replace(r'[*+-]', '', regex=True)

df
```

# Changing column data types

```python
df.info()
```

## Handling empty cells, so that they don't stop the column data type change

```python
df = df.replace('', pd.NA)
```

```python
df.isna().sum().sort_values(ascending=False)
```

```python
df = df.fillna(0)
```

## Now changing the types

```python
df = df.astype( {'Highest_Score' : int,
            'Balls_Faced' : int,
            'Strike_Rate' : float,
            '0' : int,
            '4s' : int,
            '6s' : int,
            'Debut_Year' : int,
            'Final_Year' : int} )
```

```python
df.info()
```

# Checking for some patterns before diving into proper analysis

Which players have the highest scores?

```python
df[['Player_Name', 'Highest_Score','Country']].sort_values(by='Highest_Score', ascending=False)
```

Which countries do the players come from?

```python
df['Country'].unique()
```

Which countries do most players come from?

```python
df['Country'].value_counts()
```

Which players are from Srilanka?

```python
df[df['Country'] == 'SL'][['Player_Name']]
```

Which players from Australia have the highest scores?

```python
aus_players_stat = df[ df['Country'] == 'AUS'][['Player_Name', 'Highest_Score']]

aus_players_stat.sort_values(by='Highest_Score', ascending=False)
```
