# Cleaning

## 1. Renaming columns for better understanding of audience

```python
df.columns
```

```python
df = df.rename(columns={'Mat' : 'Matches',
                        'Inns' : 'Innings',
                        'NO' : 'Not_Out',
                        'HS' : 'Highest_Score',
                        'Ave' : 'Average',
                        'BF' : 'Balls_Faced',
                        'SR' : 'Strike_Rate'})

df
```

## 2. Removing any duplicate row in the dataset

```python
df.duplicated().sum()
```

since it is zero, there's no duplicate row here. So we can move on to the next step

if there were duplicate rows, we would remove them using: *df.drop_duplicates()*

## 3. Splitting columns to make more easily readable data

```python
#checking how the data looks right now

df
```

### 1. Break the 'Player' column into 'Player_Name' and 'Country'

```python
df[['Player_Name', 'Country']] = df['Player'].str.split('(', n=2, expand=True)

df
```

```python
# remove the closing bracket from the Country column values
df['Country'] = df['Country'].str.replace(')', '')

df['Country']
```

```python
df['Country'].value_counts()
```

```python
df['Country'].unique()
```

```python
df['Country'] = df['Country'].str.replace('ICC/', '')

df['Country'].value_counts()
```

```python
#removing the redundant 'Player' column now
df = df.drop(columns='Player')

df
```

```python
# bringing the Player_Name column to the front (0 index)

temp_col_holder = df.pop('Player_Name')

df.insert(0, 'Player_Name', temp_col_holder)

df
```

### 2. Break the 'Span' column into 'Debut_Year' and 'Final_Year'

```python
df[['Debut_Year', 'Final_Year']] = df['Span'].str.split('-', n=2, expand=True)

df
```

```python
#removing the redundant Span column
df = df.drop(columns='Span')

df
```

## 4. Removing unwanted characters

```python
df
```

### Removing * and + from all columns

```python
df = df.replace(r'[+*-]', '', regex=True)
'''
By passing regex=True, you’re telling Pandas:
“Interpret the pattern I give you as a regular expression, not a literal string.”
So instead of matching exactly "+*-", it will look for any text that matches the regex pattern +*-
'''

df
```

```python
#let's check a sample of the data to see how the values look like now
df.sample(10)
```

## 5. Handle null values

```python
df.info()
```

This is still showing no null value, because it is considering the empty string in certain columns or rows AS just empty strings. We have to fix that

```python
df = df.replace('', pd.NA) #changing empty strings into the default null value of Pandas
```

```python
#running the info method again to see if it detects the null values now
df.info()
```

```python
df.isna().sum().sort_values(ascending=False)
```

### Filing up the numerical null values with 0 (zero)

```python
df = df.fillna(0)
```

```python
df.isna().sum().sort_values(ascending=False)
```

  * Filling up the numerical null values with average
  * Filling up the categorical null values with empty string

## 6. Proper column data types

```python
df.dtypes
```

These columns should be int: Highest_Score, Balls_Faced, 0, 4s, 6s, Debut_Year, Final_Year

These columns should be float: Strike_Rate

```python
df = df.astype({'Highest_Score' : int,
           'Balls_Faced' : int,
           '0' : int,
           '4s' : int,
           '6s' : int,
           'Debut_Year' : int,
           'Final_Year' : int,
           'Strike_Rate' : float})

df.dtypes
```
